```python
# train your Linear Regression model here. Data has already been split into predictors (X_train, X_val and X_test) and the
# desired response (y_train, y_val and y_test above).
#
# Note that you DO NOT NEED to remove any variables, or add higher order terms. JUST USE THE DATA AS IS.
#

# Once you've trained the model, evaluate it. This would include:
#  - Measuring the accuracy of the model (i.e. RMSE or similar measures)
#  - Looking at the properties of the fitted model, for example the R^2, and if all terms are contributing to the model
#  - Considering the validity of the trained model. This could include assumptions about independent predictors, and the 
#    distribution of the residuals
# Your write-up should include your findings regarding the above.

import statsmodels.api as sm

X_train_const = sm.add_constant(X_train)
model = sm.OLS(y_train, X_train_const).fit()

model.summary()
```

```python
X_train_const = sm.add_constant(X_train_scaled)
model_linear = sm.OLS(y_train, X_train_const).fit()

model_linear.summary()
```

```python
# train your regularised models (ridge and lasso). Here you will use the same data as you used in for the above regression model
#
# One key consideration with your regularised models is the selection of lambda (i.e. the strength of the regularisation). A suggested 
# starting point for finding your value of lambda is suggested below using numpy.logspace (see 
# https://numpy.org/doc/stable/reference/generated/numpy.logspace.html). The code below will give you 50 values starting at 10-10 and 
# finishing at 1, spread on a log scale. Just note that for plotting things that result from this, you may want to change the scale of 
# your x-axis to a log scale (see https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.xscale.html)
#
# Note that the below list of lambda is going to be fairly coarse. For the sake of speed, we suggest that you avoid evaluating a huge
# number of lambdas in one pass, but you can get a more precise estimate by refining your estimate over multiple passes (i.e. if your
# initial estimate if 0.1, you could run the search again now looking over a range of [0.01 .. 1.0] or similar). If you wish to do this,
# we suggest limiting yourself to 2-3 passes at most. You do not need to get the perfect value of lambda.
#
# remember that lambda is a keyword in python, so we're calling the array below alphas. 
#
from sklearn.linear_model import RidgeCV
from sklearn.model_selection import cross_val_score

alphas = np.logspace(-10, 0)

ridge_model = RidgeCV(alphas=alphas, store_cv_values=True)
ridge_model.fit(X_train_scaled, y_train)
# Once you've found your ridge and lasso models, evaluate them, following the approach you've used for the linear method, and compare the
# three models.

ridge_model.alpha_
```

```python
from sklearn.linear_model import LassoCV

lasso_model = LassoCV(alphas=alphas, cv=5, max_iter=10000)
lasso_model.fit(X_train_scaled, y_train)
```

```python
import numpy as np
import statsmodels.api as sm
from sklearn.metrics import mean_squared_error, mean_absolute_error
from statsmodels.stats.outliers_influence import variance_inflation_factor
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats

X_train_const = sm.add_constant(X_train)  
ols_model = sm.OLS(y_train, X_train_const).fit()
ols_pred_train = ols_model.predict(X_train_const)

ridge_pred_train = ridge_model.predict(X_train_scaled)
lasso_pred_train = lasso_model.predict(X_train_scaled)

X_val_const = sm.add_constant(X_val)  
ols_pred_val = ols_model.predict(sm.add_constant(X_val))
ols_pred_test = ols_model.predict(sm.add_constant(X_test))

ridge_pred_val = ridge_model.predict(X_val_scaled)
ridge_pred_test = ridge_model.predict(X_test_scaled)

lasso_pred_val = lasso_model.predict(X_val_scaled)
lasso_pred_test = lasso_model.predict(X_test_scaled)

def evaluate_model(y_true, y_pred, model_name):
    r2 = round(np.corrcoef(y_true, y_pred)[0, 1] ** 2, 4) 
    rmse = round(np.sqrt(mean_squared_error(y_true, y_pred)), 4)
    mae = round(mean_absolute_error(y_true, y_pred), 4)
    
    print(f"{model_name}:\n  RÂ²: {r2}\n  RMSE: {rmse}\n  MAE: {mae}\n")
    return r2, rmse, mae

print("Training Set Metrics:")
ols_results_train = evaluate_model(y_train, ols_pred_train, "OLS")
ridge_results_train = evaluate_model(y_train, ridge_pred_train, "Ridge")
lasso_results_train = evaluate_model(y_train, lasso_pred_train, "LASSO")

print("Validation Set Metrics:")
ols_results_val = evaluate_model(y_val, ols_pred_val, "OLS")
ridge_results_val = evaluate_model(y_val, ridge_pred_val, "Ridge")
lasso_results_val = evaluate_model(y_val, lasso_pred_val, "LASSO")

print("Test Set Metrics:")
ols_results_test = evaluate_model(y_test, ols_pred_test, "OLS")
ridge_results_test = evaluate_model(y_test, ridge_pred_test, "Ridge")
lasso_results_test = evaluate_model(y_test, lasso_pred_test, "LASSO")

vif_data = pd.DataFrame()
vif_data["Feature"] = X_train.columns
vif_data["VIF"] = [variance_inflation_factor(X_train_const, i) for i in range(1, X_train_const.shape[1])]

print("\nMulticollinearity Check (VIF):")
print(vif_data)

ols_residuals_train = y_train - ols_pred_train

plt.figure(figsize=(6, 4))
stats.probplot(ols_residuals_train, dist="norm", plot=plt)
plt.title("Q-Q Plot of OLS Residuals")
plt.show()

plt.figure(figsize=(6, 4))
plt.hist(ols_residuals_train, bins=30, edgecolor='black')
plt.title("Histogram of Residuals (OLS)")
plt.xlabel("Residual")
plt.ylabel("Frequency")
plt.show()

ridge_residuals_train = y_train - ridge_pred_train

plt.figure(figsize=(6, 4))
stats.probplot(ridge_residuals_train, dist="norm", plot=plt)
plt.title("Q-Q Plot of Ridge Residuals")
plt.show()

plt.figure(figsize=(6, 4))
plt.hist(ridge_residuals_train, bins=30, edgecolor='black')
plt.title("Histogram of Residuals (Ridge)")
plt.xlabel("Residual")
plt.ylabel("Frequency")
plt.show()

lasso_residuals_train = y_train - lasso_pred_train

plt.figure(figsize=(6, 4))
stats.probplot(lasso_residuals_train, dist="norm", plot=plt)
plt.title("Q-Q Plot of Lasso Residuals")
plt.show()

plt.figure(figsize=(6, 4))
plt.hist(lasso_residuals_train, bins=30, edgecolor='black')
plt.title("Histogram of Residuals (Lasso)")
plt.xlabel("Residual")
plt.ylabel("Frequency")
plt.show()

ols_residuals_test = y_test - ols_pred_test
ridge_residuals_test = y_test - ridge_pred_test
lasso_residuals_test = y_test - lasso_pred_test

plt.figure(figsize=(6, 4))
stats.probplot(ols_residuals_test, dist="norm", plot=plt)
plt.title("Q-Q Plot of OLS Residuals (Test Set)")
plt.show()

plt.figure(figsize=(6, 4))
stats.probplot(ridge_residuals_test, dist="norm", plot=plt)
plt.title("Q-Q Plot of Ridge Residuals (Test Set)")
plt.show()

plt.figure(figsize=(6, 4))
stats.probplot(lasso_residuals_test, dist="norm", plot=plt)
plt.title("Q-Q Plot of Lasso Residuals (Test Set)")
plt.show()

plt.figure(figsize=(6, 4))
plt.hist(ols_residuals_test, bins=30, edgecolor='black')
plt.title("Histogram of Residuals (OLS - Test Set)")
plt.xlabel("Residual")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(6, 4))
plt.hist(ridge_residuals_test, bins=30, edgecolor='black')
plt.title("Histogram of Residuals (Ridge - Test Set)")
plt.xlabel("Residual")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(6, 4))
plt.hist(lasso_residuals_test, bins=30, edgecolor='black')
plt.title("Histogram of Residuals (Lasso - Test Set)")
plt.xlabel("Residual")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(6, 4))
plt.scatter(y_test, ols_pred_test, edgecolor='black')
plt.title("Predicted vs Actual (OLS - Test Set)")
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.show()

plt.figure(figsize=(6, 4))
plt.scatter(y_test, ridge_pred_test, edgecolor='black')
plt.title("Predicted vs Actual (Ridge - Test Set)")
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.show()

plt.figure(figsize=(6, 4))
plt.scatter(y_test, lasso_pred_test, edgecolor='black')
plt.title("Predicted vs Actual (Lasso - Test Set)")
plt.xlabel("Actual Values")
plt.ylabel("Predicted Values")
plt.show()

plt.figure(figsize=(10, 6))
plt.plot(y_test.values, label="Actual", color='blue', linestyle='-', marker='o', markersize=4)
plt.plot(ols_pred_test, label="OLS Predicted", color='red', linestyle='-', marker='x', markersize=4)
plt.plot(ridge_pred_test, label="Ridge Predicted", color='green', linestyle='-', marker='x', markersize=4)
plt.plot(lasso_pred_test, label="Lasso Predicted", color='orange', linestyle='-', marker='x', markersize=4)
plt.title("Actual vs Predicted Values (Test Set)")
plt.xlabel("Index")
plt.ylabel("Target Variable")
plt.legend()
plt.show()
```
